```python
# ============================================================
# EXPORT CELL — fully self-contained
#
# Make sure draw_face_helpers/geometry.py is the patched version,
# __pycache__ is cleared, and kernel is restarted.
# ============================================================

import os
import numpy as np
import colorsys
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image

from draw_face_helpers import (
    draw_face_patch, draw_face_brain, add_circle_lens_shadow,
    add_one_brain_circuit_from_brain_top,
    add_rect_bent_double_node_from_brain_top,
    draw_eyes, add_smile, background_circle,
)

# ═══════════════════════════════════════════════════════════════
# Constants
# ═══════════════════════════════════════════════════════════════
DPI = 140
LW_OUTLINE = 11
REF_PX = 840

FACE_SIZE = 0.86
FACE_ROUND_TOP = 0.44
FACE_ROUND_BOT = 0.44

BRAIN_COLOR = "#EEF2F8"
FACE_COLOR  = "#FFFFFF"
LINE_COLOR  = "#0B0F14"
BACKGROUND_COLOR = "#7FE6D4"

WIRE_COLOR = LINE_COLOR
NODE_LED   = "#26D3B3"
NODE_METAL = "#334155"
NODE_EDGE  = LINE_COLOR

EYE_IRIS          = NODE_LED
EYE_PUPIL_RATIO   = 0.35
EYE_SPEC_RATIO    = 0.22
EYE_GAZE_STRENGTH = 0.45

EYES_CY = 0.42
SMILE_Y = -0.14
SMILE_W = 0.36
SMILE_H = 0.20

SHADOW_Z  = 28
CIRCUIT_Z = 40

# ── Color helpers ────────────────────────────────────────────
def _hex_to_rgb01(h):
    h = h.lstrip("#")
    return tuple(int(h[i:i+2], 16)/255.0 for i in (0, 2, 4))

def _rgb01_to_hex(rgb):
    return "#{:02X}{:02X}{:02X}".format(
        int(round(rgb[0]*255)), int(round(rgb[1]*255)), int(round(rgb[2]*255)))

def derive_cold_tint(base_hex, hue_shift=-0.06, sat_floor=0.35, val_floor=0.75, val_scale=0.92):
    r, g, b = _hex_to_rgb01(base_hex)
    h, s, v = colorsys.rgb_to_hsv(r, g, b)
    s = max(s, sat_floor)
    h = (h + hue_shift) % 1.0
    v = max(v * val_scale, val_floor)
    return _rgb01_to_hex(colorsys.hsv_to_rgb(h, s, v))

def cold_shadow_palette(base_hex, n=1, sat_min=0.08, sat_max=0.14, val_min=0.97, val_max=0.90):
    r, g, b = _hex_to_rgb01(base_hex)
    h, s, v = colorsys.rgb_to_hsv(r, g, b)
    return [_rgb01_to_hex(colorsys.hsv_to_rgb(h, ss, vv))
            for ss, vv in zip(np.linspace(sat_min, sat_max, n),
                              np.linspace(val_min, val_max, n))]

COLD_TINT_FACE  = derive_cold_tint(FACE_COLOR,  hue_shift=-0.07, sat_floor=0.30, val_floor=0.72, val_scale=0.90)
COLD_TINT_BRAIN = derive_cold_tint(BRAIN_COLOR, hue_shift=+0.10, sat_floor=0.45, val_floor=0.68, val_scale=0.88)
face_band_colors  = cold_shadow_palette(COLD_TINT_FACE,  n=1)
brain_band_colors = cold_shadow_palette(COLD_TINT_BRAIN, n=1, sat_min=0.10, sat_max=0.18, val_min=0.96, val_max=0.86)

def add_lens_shadow_subdivisions_with_colors(ax, face_outline, center_start, center_end,
                                              radius, face_colors, brain_colors,
                                              region="outside", prefer="right", zorder=28, alpha=1.0):
    n = len(face_colors)
    if n == 0: return
    c0, c1 = np.array(center_start, dtype=float), np.array(center_end, dtype=float)
    centers = [tuple((c0 + c1) / 2.0)] if n == 1 else [tuple(c0*(1-t) + c1*t) for t in np.linspace(0,1,n)]
    for cc, fc, bc in zip(centers, face_colors, brain_colors):
        add_circle_lens_shadow(ax, face_outline=face_outline, circle_center=cc, radius=radius,
                               brain_shadow_color=bc, face_shadow_color=fc,
                               alpha=alpha, region=region, prefer=prefer, zorder=zorder)

# ═══════════════════════════════════════════════════════════════
# Export config
# ═══════════════════════════════════════════════════════════════
TARGETS = {
    "favicon-16":  16,  "favicon-32":  32,  "favicon-48": 48,
    "logo-64":     64,  "logo-128":   128,
    "apple-touch-icon-180": 180,
    "android-192": 192, "logo-256":   256,
    "logo-512":    512, "android-512": 512,
    "logo-1024":  1024,
}
ICO_SIZES = [16, 32, 48]
OUT_DIR = "../app/assets/images"



# ═══════════════════════════════════════════════════════════════
# Render function
# ═══════════════════════════════════════════════════════════════
def render_logo(target_px, simplify_below=64):
    scale = target_px / REF_PX
    is_tiny = target_px < simplify_below

    dpi = DPI
    figsize = target_px / dpi

    # Scale point-based strokes
    lw       = LW_OUTLINE * scale
    bg_lw    = 6 * scale
    wire_lw  = lw * 0.50
    wink_lw  = lw * 0.75
    smile_lw = lw * 0.80

    if is_tiny:
        lw       = max(lw, 1.8)
        bg_lw    = max(bg_lw, 1.0)
        wire_lw  = max(wire_lw, 1.0)
        wink_lw  = max(wink_lw, 1.2)
        smile_lw = max(smile_lw, 1.2)

    fig, ax = plt.subplots(figsize=(figsize, figsize), dpi=dpi)

    # ── Transparent figure background ────────────────────────
    fig.patch.set_facecolor("none")
    fig.patch.set_alpha(0.0)
    ax.set_facecolor("none")

    ax.set_aspect("equal"); ax.axis("off")
    fig.canvas.draw()

    # Face outline
    fo = draw_face_patch(
        width=FACE_SIZE, height=FACE_SIZE, angle_top_to_bottom=0,
        rounding_top=FACE_ROUND_TOP, rounding_bottom=FACE_ROUND_BOT,
        cx=0.5, cy=0.5, facecolor="none", edgecolor=LINE_COLOR, lw=lw, zorder=20)
    ax.add_patch(fo)

    # Brain/face fills
    draw_face_brain(
        ax, fo, cx=0.5, cy=0.5, y_pos_px=30, curvature_px=-55,
        curve_dir="down", fill_pad_px=0, sep_n=700, boundary_interp=220,
        lw=lw, brain_color=BRAIN_COLOR, face_color=FACE_COLOR, line_color=LINE_COLOR)

    # Lens shadow — skip tiny
    if not is_tiny:
        add_lens_shadow_subdivisions_with_colors(
            ax, fo, center_start=(0.5-1.22, 0.5), center_end=(0.5-1.20, 0.5),
            radius=1.52, face_colors=face_band_colors, brain_colors=brain_band_colors,
            alpha=0.035, zorder=SHADOW_Z)

    # Circuits — skip tiny
    if not is_tiny:
        for shift in [-0.02, -0.30]:
            add_one_brain_circuit_from_brain_top(
                ax, fo, center_shift=shift, length_px=80, node_r=0.018,
                node_facecolor="#FFFFFF", node_edgecolor=NODE_EDGE,
                wire_lw=wire_lw, wire_color=WIRE_COLOR, zorder=CIRCUIT_Z)
        for shift, bdir in [(-0.02, -1), (0.20, +1)]:
            add_rect_bent_double_node_from_brain_top(
                ax, fo, center_shift=shift, length_px=75, bend_px=40,
                bend_at=0.7, bend_dir=bdir, corner_px=12, corner_n=7,
                node1_on="vertical1", node1_u=0.58,
                node_facecolor1=NODE_LED, node_facecolor2=NODE_METAL,
                node_edgecolor=NODE_EDGE, node_r=0.018,
                wire_color=WIRE_COLOR, wire_lw=wire_lw, zorder=CIRCUIT_Z)

    # Eyes
    draw_eyes(
        ax, fo, cx=0.5, cy=EYES_CY, open_left=False, open_right=True,
        eye_dx=0.19, eye_dy=-0.02, r=0.066, wink_lw=wink_lw,
        outline_color=LINE_COLOR, iris_color_left=EYE_IRIS, iris_color_right=EYE_IRIS,
        ring_ratio=0.10, iris_ratio=0.38, pupil_ratio=EYE_PUPIL_RATIO,
        spec_ratio=EYE_SPEC_RATIO, spec_offset=(0.30, 0.34), iris_offset_y=0.0,
        gaze_left=(1.0, 0.0), gaze_right=(-1.0, 0.0),
        gaze_strength=EYE_GAZE_STRENGTH, aa_shrink=0.995)

    # Smile
    add_smile(
        ax, fo, cx=0.5, cy=0.5, y=SMILE_Y, w=SMILE_W, h=SMILE_H,
        theta1=210, theta2=330, lw=smile_lw, color=LINE_COLOR)

    # Background circle
    background_circle(
        ax, center=(0.5, 0.5), radius=0.60, color=BACKGROUND_COLOR,
        alpha=0.90, edgecolor="#0B0F14", lw=bg_lw, zorder=0)

    # Redraw outline on top
    top = draw_face_patch(
        width=FACE_SIZE, height=FACE_SIZE, angle_top_to_bottom=0,
        rounding_top=FACE_ROUND_TOP, rounding_bottom=FACE_ROUND_BOT,
        cx=0.5, cy=0.5, facecolor="none", edgecolor=LINE_COLOR, lw=lw, zorder=1000)
    ax.add_patch(top)

    # Set tight limits so the background circle fills the canvas edge-to-edge
    # Circle is at (0.5, 0.5) with radius 0.60 + half the outline stroke
    pad = 0.02  # tiny breathing room
    ax.set_xlim(0.5 - 0.60 - pad, 0.5 + 0.60 + pad)
    ax.set_ylim(0.5 - 0.60 - pad, 0.5 + 0.60 + pad)
    return fig


def fig_to_pil(fig, target_px):
    buf = BytesIO()
    fig.savefig(buf, format="png",
                bbox_inches="tight",
                pad_inches=0,           # no padding
                transparent=True)       # transparent background
    plt.close(fig)
    buf.seek(0)
    img = Image.open(buf).convert("RGBA")
    if img.size != (target_px, target_px):
        img = img.resize((target_px, target_px), Image.LANCZOS)
    return img


# ═══════════════════════════════════════════════════════════════
# Run export
# ═══════════════════════════════════════════════════════════════
os.makedirs(OUT_DIR, exist_ok=True)
images = {}

for name, px in sorted(TARGETS.items(), key=lambda x: x[1]):
    print(f"  {name} ({px}×{px}) …")
    fig = render_logo(px)
    img = fig_to_pil(fig, px)
    img.save(os.path.join(OUT_DIR, f"{name}.png"))
    images[px] = img

# Build favicon.ico
ico_imgs = [images[s] for s in ICO_SIZES if s in images]
if ico_imgs:
    ico_imgs[0].save(
        os.path.join(OUT_DIR, "favicon.ico"), format="ICO",
        sizes=[(s, s) for s in ICO_SIZES], append_images=ico_imgs[1:])
    print(f"  favicon.ico ({', '.join(map(str, ICO_SIZES))} px)")

# ── Google Workspace logo (320×320 with white background) ────
print("  google-workspace-320 (320×320, white bg) …")
fig_gw = render_logo(320)
img_gw = fig_to_pil(fig_gw, 320)

# Flatten onto white background
white_bg = Image.new("RGBA", img_gw.size, (255, 255, 255, 255))
white_bg.paste(img_gw, (0, 0), img_gw)  # use alpha as mask
white_bg.convert("RGB").save(os.path.join(OUT_DIR, "google-workspace-320.png"))

# ── Also make opaque versions for any context that needs them ─
for label, px in [("logo-512-opaque", 512), ("logo-256-opaque", 256)]:
    if px in images:
        print(f"  {label} ({px}×{px}, white bg) …")
        bg = Image.new("RGBA", (px, px), (255, 255, 255, 255))
        bg.paste(images[px], (0, 0), images[px])
        bg.convert("RGB").save(os.path.join(OUT_DIR, f"{label}.png"))

print(f"\nDone → {OUT_DIR}/")
```
